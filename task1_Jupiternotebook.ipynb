```python
import cv2 as cv
```

```python
img= cv.imread('fuc.jpg')
grayimg=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
edges= cv.Canny(grayimg, 100, 200)
contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
```

```python
list = []
for i in contours:
    area= cv.contourArea(i)
    list.append(area)
```

```python
x=sorted(contours, key=cv.contourArea, reverse=True)
Big=x[0]
cv.drawContours(img, Big, -1, (255,0,0), 7)
```
